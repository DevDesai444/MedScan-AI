```python
# Install required packages if you haven’t already
# pip install transformers accelerate peft torch torchvision datasets pillow

from transformers import AutoProcessor, VisionEncoderDecoderModel
import os

# 1. Specify the HF model repo
MODEL_NAME = "inflectionpoint/qwen-2-3b-vl"

# 2. Download and cache the processor
print("Downloading processor…")
processor = AutoProcessor.from_pretrained(MODEL_NAME)

# 3. Download and cache the vision-language model
print("Downloading VisionEncoderDecoderModel…")
model = VisionEncoderDecoderModel.from_pretrained(MODEL_NAME)

# 4. Verify download
print("Model and processor loaded:")
print(" - Processor:", processor)
print(" - Model:", model)

# 5. (Optional) Save locally for offline use
CACHE_DIR = "./cache/qwen2_3b_vl"
os.makedirs(CACHE_DIR, exist_ok=True)
processor.save_pretrained(CACHE_DIR)
model.save_pretrained(CACHE_DIR)
print(f"Saved processor and model to {CACHE_DIR}")
```
